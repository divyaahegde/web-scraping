```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
path = "https://www.amazon.in/Test-Exclusive-549/product-reviews/B077PWBC78/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
data = requests.get(path)
data
###re-run the code cell if you get a response other than <Response [200]>
```

```python
data.content
```

```python
soup = bs(data.content,'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')[2:]
names
```

```python
cust_name = []
for i in range(0,len(names)):
  cust_name.append(names[i].get_text())
cust_name
```

```python
ratings = soup.find_all('span',class_='a-icon-alt')[3:]
len(ratings)
```

```python
cust_rating = []
for i in range(0,len(ratings)):
  cust_rating.append(ratings[i].get_text())
cust_rating
```

```python
title = soup.find_all('a',{"data-hook":"review-title"})
title
```

```python
review_title = []
for i in range(0,len(title)):
  review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [a.strip('\n') for a in review_title]
review_title
```

```python
date = soup.find_all('span',{"data-hook":"review-date"})
date
```

```python
review_date = []
for i in range(0,len(date)):
  review_date.append(date[i].get_text())
review_date
```

```python
content = soup.find_all('span',{"data-hook":"review-body"})
content
```

```python
review_body = []
for i in range(0,len(content)):
  review_body.append(content[i].get_text())
review_body
```

```python
review_body[:] = [a.strip('\n') for a in review_body]
review_body
```

```python
import pandas as pd
df = pd.DataFrame()
df['Date'] = review_date
df['Customer Name'] = cust_name
df['Review Title'] = review_title
df['Ratings'] = cust_rating
df['Review Body'] = review_body
df
```

```python
df.to_csv('amazon.csv')
```
